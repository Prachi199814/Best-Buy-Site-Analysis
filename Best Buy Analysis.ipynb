```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Data.csv')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
d=df[['SKU','Number of Reviews','Rating','Price']]
df1=pd.DataFrame(d)
corr_matrix=df1.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
Count_by_Brand=df.groupby(by='Brand').agg(Total_Products=('Serial Number','count')).sort_values(by='Total_Products',ascending=False).head(15)
```

```python
Count_by_Brand
```

```python
plt.figure(figsize=(14,6))
plt.title('Total number of Products by Brand')
sns.barplot(x='Brand',y='Total_Products',data=Count_by_Brand,palette='coolwarm')
plt.show()
```

```python
Average_Price_Df=df.groupby(by='Category').agg(Average_Price=('Price','mean')).sort_values(by='Average_Price',ascending=False).head(15)
```

```python
Average_Price_Df
```

```python
plt.figure(figsize=(14,6))
plt.title('Average Price by Category')
sns.barplot(x='Category',y='Average_Price',data=Average_Price_Df,palette='viridis')
plt.show()
```

```python
Max_Priced_Df=df.groupby(by='Category')['Price'].idxmax()
```

```python
Max_Priced=df.loc[Max_Priced_Df,['Category','Price','Name']]
```

```python
Max_Priced
```

```python
plt.figure(figsize=(14,6))
plt.title('Maximum Price by Category')
sns.lineplot(x='Category',y='Price',data=Max_Priced,palette='twilight')
plt.show()
```

```python
Rating_By_Brand=df.groupby(by='Brand').agg(Average_Rating=('Rating','mean')).sort_values(by='Average_Rating',ascending=False).head(10)
```

```python
Rating_By_Brand
```

```python
df['Sum of Ratings']=df['Rating']*df['Number of Reviews']
```

```python
df
```

```python
Ratings_By_Brand=df.groupby(by='Brand').agg(Sum_of_Ratings=('Sum of Ratings','sum'),Total_Reviews=('Number of Reviews','sum'))
```

```python
Ratings_By_Brand
```

```python
Ratings_By_Brand['Calculated Rating']=Ratings_By_Brand['Sum_of_Ratings']/Ratings_By_Brand['Total_Reviews']
```

```python
Ratings_By_Brand.sort_values(by='Calculated Rating',ascending=False).head(15)
```

```python
Rating_By_Brand
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='Brand',y='Average_Rating',data=Rating_By_Brand,hue='Brand',palette='deep')
plt.show()
```

```python
Ratings_By_Brand=df.groupby(by='Color or Quality').agg(Total_Numbers=('Name','count')).sort_values(by='Total_Numbers',ascending=False).head(10)
```

```python
Ratings_By_Brand
```

```python
Ratings_By_Brand['Total_Numbers'].plot(kind='barh',color='red')
plt.title('Number of Products by Color or Quality')
plt.xlabel('Total Numbers')
plt.show()
```
