```python
import pandas as pd
```

```python
df=pd.read_csv('Best Buy Data.csv')
```

```python
df
```

```python
df['Serial Number']=df['Unnamed: 0']
```

```python
df.drop('Unnamed: 0',axis='columns',inplace=True)
```

```python
df
```

```python
df.set_index(['Serial Number'],inplace=True)
```

```python
df
```

```python
Brands=[]
for x in df['Name']:
    Brands.append(x.split(' ')[0])
```

```python
df['Brand']=Brands
```

```python
df
```

```python
Color_or_Quality=[]
for x in df['Name']:
    Color_or_Quality.append(x.split('-')[-1])
```

```python
df['Color or Quality']=Color_or_Quality
```

```python
df
```

```python
df.info()
```

```python
df['Number of Reviews'].replace('Ye Revewed',0,inplace=True)
```

```python
df['Number of Reviews']=pd.to_numeric(df['Number of Reviews'])
```

```python
df.info()
```

```python
df['Rating'].replace('Nt Yet Re',0,inplace=True)
```

```python
df['Rating']=pd.to_numeric(df['Rating'])
```

```python
df['Price']=pd.to_numeric(df['Rating'])
```
